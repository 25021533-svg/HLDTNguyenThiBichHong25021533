```python
#Kiểm tra có phải số nguyên tố hay không
import math
def checkprime_number(n:int)->bool:
    if n<2:
        return False
    for i in range(2,math.isqrt(n)+1):
        if n%i==0:
            return False    
    return True
n=int(input("Nhập số nguyên n: "))
print(checkprime_number(n))
```

```python
import math

# Bước 1: Định nghĩa hàm kiểm tra số nguyên tố
def is_prime(num):
    if num < 2:
        return False
    # Chỉ cần kiểm tra đến căn bậc hai của num để tối ưu
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            return False
    return True

# Bước 2: Nhập dữ liệu
n = int(input("Nhập số nguyên n: "))

# Bước 3: Duyệt và in các số nguyên tố nhỏ hơn n
print(f"Các số nguyên tố nhỏ hơn {n} là:")
for i in range(2, n):
    if is_prime(i):
        print(i, end=" ")
```

```python
#Nhập vào một số nguyên dương, tính tổng các chữ số nguyên tố của nó
n=int(input())
def sum_of_prime_digits_numbers(n:int) -> int:
    tổng = 0
    while n > 0:
        digit = n % 10
        if checkprime_number(digit):
            tổng += digit
        n //= 10
    return tổng
print(sum_of_prime_digits_numbers(n))
```
