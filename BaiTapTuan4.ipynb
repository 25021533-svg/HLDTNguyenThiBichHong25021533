```python
#W4A2. PrimeNumber: Nhập cho đến khi nhập số dương, kiểm tra số nguyên tố
while True:
    n = int(input("Nhập n: "))
    if n > 0:
        break #Nếu n thoả mãn n>0 thì thoát khỏi vòng lặp, bước sang phần kiểm tra số nguyên tố
is_prime = True #Giả sử n là số nguyên tố 
if n == 1:
    is_prime = False #Nếu n=1 thì n không phải số nguyên tố
else:
    i = 2 #Bắt đầu kiểm tra từ 2 đến căn bậc hai của n, nghĩa là i^2<=n
    while i*i <= n: 
        if n % i == 0: #Nếu n chia hết cho i thì n không phải số nguyên tố
            is_prime = False #Khi n không phải số nguyên tố thì gán is_prime=False
            break #Thoái khỏi vòng lặp
        i += 1 #Nghĩa là i+1=i để kiểm tra số tiếp theo
print("Prime" if is_prime else "Not prime")
```

```python
#W4A3. Factorial: Giai thừa n (0 < n < 100)
n = int(input())
g = 1
for i in range(1, n+1):
    g *= i #g=g*i, vòng lặp for nghĩa là g_1=1*1, g_2=1*2, g_3=2*3, g_4=6*4,... => g
print(g)
```

```python
#W4A4. CountingDigits: Không dùng string
n = int(input())
if n == 0: 
    print(1) 
else:
    n = abs(n) #Lấy giá trị tuyệt đối của n để tránh trường hợp n âm (abs(-123)=123)
    count = 0
    while n > 0:
        count += 1 #Nghĩa là đếm số kí tự, count=count+1
        n //= 10 #Nghĩa là n=n//10, dùng để loại bỏ chữ số cuối cùng của n sau mỗi lần lặp 
    print(count)
```

```python
#W4A6. SumPrimesInRange: Tổng số nguyên tố trong đoạn [a, b]
a, b = map(int, input().split())
def is_prime(n):
    if n < 2:
        return False
    i = 2
    while i*i <= n:
        if n % i == 0:
            return False
        i += 1
    return True
s = 0
for x in range(a, b+1):
    if is_prime(x):
        s += x
print(s)
```

```python
#W4A7. LargestPrimeFactor: Ước số nguyên tố lớn nhất của n
n = int(input())
i = 2
ans = 1
while i*i <= n:
    while n % i == 0:
        ans = i
        n //= i
    i += 1
if n > 1:
    ans = n
print(ans)
```

```python
#W4A8. ReverseAddPalindrome: Lặp n = n + reverse(n) cho đến khi thành palindrome
n = int(input())
def reverse(x):
    r = 0
    while x > 0:
        r = r*10 + x%10
        x //= 10
    return r
def is_pal(x):
    return x == reverse(x)
count = 0
while not is_pal(n):
    n = n + reverse(n)
    count += 1
print(count, n)
```

```python
#W4A9. PerfectSquaresDistinctDigits: Các số chính phương ≤ n có chữ số đôi một khác nhau
n = int(input())
found = False
def all_distinct(x):
    seen = [0]*10
    while x > 0:
        d = x % 10
        if seen[d] == 1:
            return False
        seen[d] = 1
        x //= 10
    return True
i = 1
while i*i <= n:
    sq = i*i
    if all_distinct(sq):
        print(sq, end=" ")
        found = True
    i += 1
if not found:
    print("no number")
```

```python
#W4A10. LongestCollatzUpToN: Tìm số có độ dài Collatz lớn nhất trong [1, n]
n = int(input())
def collatz_length(x):
    length = 1
    while x != 1:
        if x % 2 == 0:
            x //= 2
        else:
            x = x*3 + 1
        length += 1
    return length
best_x = 1
best_len = 1
for x in range(1, n+1):
    L = collatz_length(x)
    if L > best_len:
        best_len = L
        best_x = x
print(best_x, best_len)
```

```python
#W4A11. EvenDivisors: Đếm ước số chẵn của n
n = int(input())
count = 0
i = 1
while i*i <= n:
    if n % i == 0:
        if i % 2 == 0:
            count += 1
        if i != n//i and (n//i) % 2 == 0:
            count += 1
    i += 1
print(count)
```

```python
#W4A13. ClosePairNumber (Cặp số thân thiết)
a = int(input())
b = int(input())
def sum_div(n):
    s = 0
    i = 1
    while i*i <= n:
        if n % i == 0:
            if i != n:
                s += i
            j = n//i
            if j != i and j != n:
                s += j
        i += 1
    return s
print("true" if sum_div(a) == b and sum_div(b) == a else "false")
```

```python
#W4A14. GCD: UCLN của m và n
m, n = map(int, input().split())
while n != 0:
    m, n = n, m % n
print(m)
```

```python
#W4A15. ChickenAndDog: Bài toán gà chó: tổng số con, tổng số chân
animals = int(input())
legs = int(input())
found = False
for ga in range(animals+1):
    cho = animals - ga
    if ga*2 + cho*4 == legs:
        print(ga, cho)
        found = True
        break
if not found:
    print("invalid")
```
