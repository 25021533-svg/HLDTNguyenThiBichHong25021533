```python
#W4A2. PrimeNumber
Nhập cho đến khi nhập số dương, kiểm tra số nguyên tố
while True:
    n = int(input("Nhập n: "))
    if n > 0:
        break
is_prime = True
if n == 1:
    is_prime = False
else:
    i = 2
    while i*i <= n:
        if n % i == 0:
            is_prime = False
            break
        i += 1
print("Prime" if is_prime else "Not prime")
```

```python
#W4A6. SumPrimesInRange
Tổng số nguyên tố trong đoạn [a, b]
a, b = map(int, input().split())

def is_prime(n):
    if n < 2:
        return False
    i = 2
    while i*i <= n:
        if n % i == 0:
            return False
        i += 1
    return True

s = 0
for x in range(a, b+1):
    if is_prime(x):
        s += x

print(s)
```

```python
#W4A7. LargestPrimeFactor
Ước số nguyên tố lớn nhất của n
n = int(input())
i = 2
ans = 1
while i*i <= n:
    while n % i == 0:
        ans = i
        n //= i
    i += 1
if n > 1:
    ans = n
print(ans)
```

```python
#W4A8. ReverseAddPalindrome
Lặp n = n + reverse(n) cho đến khi thành palindrome
n = int(input())
def reverse(x):
    r = 0
    while x > 0:
        r = r*10 + x%10
        x //= 10
    return r
def is_pal(x):
    return x == reverse(x)
count = 0
while not is_pal(n):
    n = n + reverse(n)
    count += 1
print(count, n)
```

```python
#W4A9. PerfectSquaresDistinctDigits
Các số chính phương ≤ n có chữ số đôi một khác nhau
n = int(input())
found = False
def all_distinct(x):
    seen = [0]*10
    while x > 0:
        d = x % 10
        if seen[d] == 1:
            return False
        seen[d] = 1
        x //= 10
    return True
i = 1
while i*i <= n:
    sq = i*i
    if all_distinct(sq):
        print(sq, end=" ")
        found = True
    i += 1
if not found:
    print("no number")
```
