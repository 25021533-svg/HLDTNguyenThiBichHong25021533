```python
#W7A1: Viết hàm tìm kiếm nhị phân trên một danh sách đã được sắp xếp tăng dần. Trả về chỉ số của phần tử hoặc -1 nếu không tìm thấy. mẫu hàm: def binary_search(arr, left, right, target)
def binary_search(arr, left, right, target):
    if left > right:
        return -1
    mid = (left + right) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] > target:
        return binary_search(arr, left, mid - 1, target)
    else:
        return binary_search(arr, mid + 1, right, target)
arr = [1, 3, 5, 7, 9, 11, 13, 15]
left = 0
right = len(arr) - 1
target = 7
print(binary_search(arr, left, right, target))
```

```python
#W7A3: Viết hàm thực hiện thuật toán Bubble Sort và Selection Sort để sắp xếp danh sách theo thứ tự tăng dần.
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
arr = [64, 34, 25, 12, 22, 11, 90]
print("Bubble Sort:", bubble_sort(arr.copy()))
print("Selection Sort:", selection_sort(arr.copy()))
```

```python
#W7A4: Tìm phần tử có số lần xuất hiện nhiều nhất. Nếu có nhiều phần tử cùng số lần, trả về phần tử xuất hiện trước.
def most_frequent(arr):
    counts = {}
    max_count = 0
    result = arr[0]
    for x in arr:
        counts[x] = counts.get(x, 0) + 1
        if counts[x] > max_count:
            max_count = counts[x]
            result = x
    return f"{result} xuat hien nhieu nhat, som nhat, {max_count} lan"
arr = [1, 3, 2, 3, 4, 1, 3, 2]
print(most_frequent(arr))
```

```python
#W7A5: Cho danh sách các số nguyên và số $X$, đếm tất cả các cặp $(a, b)$ sao cho $a+b=X$.
def count_pairs_with_sum(arr, X):
    count = 0
    seen = {}
    for x in arr:
        target = X - x
        if target in seen:
            count += seen[target]
        seen[x] = seen.get(x, 0) + 1
    return count

arr = [1, 2, 3, 4, 3, 2, 1]
X = 5
print(count_pairs_with_sum(arr, X))
```

```python
#W7A6:Cung cấp một danh sách các danh sách lồng nhau, hãy làm phẳng chúng.
def flatten_list(nested_list):
    flat_list = []
    for sublist in nested_list:
        for item in sublist:
            flat_list.append(item)
    return flat_list
nested_list = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
print(flatten_list(nested_list))
```

```python
#W7A7:Tìm độ dài của dãy con tăng nghiêm ngặt dài nhất (không nhất thiết phải liền kề).
import bisect

def length_of_lis(nums):
    if not nums: return 0
    tails = []
    for x in nums:
        idx = bisect.bisect_left(tails, x)
        if idx < len(tails):
            tails[idx] = x
        else:
            tails.append(x)
    return len(tails)
nums = [10, 9, 2, 5, 3, 7, 101, 18]
print(length_of_lis(nums))
```

```python
#W7A8:
import heapq

def shortest_interval(intervals, queries):
    # Sắp xếp intervals theo l_i, queries theo giá trị
    sorted_queries = sorted([(q, i) for i, q in enumerate(queries)])
    intervals.sort()
    
    res = [-1] * len(queries)
    pq = [] # (độ dài, r_i)
    curr_int = 0
    
    for q_val, original_idx in sorted_queries:
        while curr_int < len(intervals) and intervals[curr_int][0] <= q_val:
            l, r = intervals[curr_int]
            heapq.heappush(pq, (r - l + 1, r))
            curr_int += 1
        
        while pq and pq[0][1] < q_val:
            heapq.heappop(pq)
            
        if pq:
            res[original_idx] = pq[0][0]
            
    return res
intervals = [[1, 4], [2, 4], [3, 6], [4, 4]]
queries = [2, 3, 4, 5]
print(shortest_interval(intervals, queries))
```

```python
#W7A10:
import bisect

def max_subarray_mod(nums, m):
    prefix_sum = 0
    sorted_prefix = [0]
    max_mod = 0
    for x in nums:
        prefix_sum = (prefix_sum + x) % m
        # Tìm prefix_sum nhỏ nhất lớn hơn hiện tại
        idx = bisect.bisect_right(sorted_prefix, prefix_sum)
        if idx < len(sorted_prefix):
            max_mod = max(max_mod, (prefix_sum - sorted_prefix[idx] + m) % m)
        else:
            max_mod = max(max_mod, prefix_sum)
        bisect.insort(sorted_prefix, prefix_sum)
    return max_mod
nums = [3, 3, 9, 9, 5]
m = 7
print(max_subarray_mod(nums, m))
```

```python
#W7A13:
def longest_harmonious_subsequence(arr):
    counts = {}
    for x in arr:
        counts[x] = counts.get(x, 0) + 1
    
    max_len = 0
    for x in counts:
        if x + 1 in counts:
            max_len = max(max_len, counts[x] + counts[x+1])
    return max_len
arr = [1, 3, 2, 2, 5, 2, 3, 7]
print(longest_harmonious_subsequence(arr))
```

```python
#W7A14:
def find_sevens(n, arr):
    indices = [str(i) for i, x in enumerate(arr) if x == 7]
    if not indices:
        return "Not found"
    return " ".join(reversed(indices))
n = 5
arr = [1, 3, 2, 2, 5, 2, 3, 7]
print(find_sevens(n, arr))
```
