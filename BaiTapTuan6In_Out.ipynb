```python
#W6A1: Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
# Nhập một dòng gồm các số nguyên
nums = list(map(int, input().split()))
result = [] # Danh sách để lưu kết quả
seen = set() # Tập để theo dõi các số đã gặp
for x in nums:
    if x not in seen: 
        result.append(x)
        seen.add(x) # Nếu số chưa gặp thì thêm vào kết quả và đánh dấu là đã gặp

print(result)
```

```python
#W6A2: Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị trí i), giữ nguyên thứ tự.
nums = list(map(int, input().split()))
cumulative_sum = [] # Danh sách để lưu tổng tích lũy
current_sum = 0 # Biến để theo dõi tổng hiện tại
for x in nums:
    current_sum += x
    cumulative_sum.append(current_sum) # Cập nhật tổng và thêm vào danh sách
print(cumulative_sum) 
```

```python
#W6A3: Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp xếp.)
nums=tuple(map(int, input().split()))
k=int(input())
k = k % len(nums)  # Đảm bảo k không vượt quá len(tuple), nếu k chia hết cho độ dài thì không cần xoay, nếu k chia có phần dư thì lấy phần dư để xoay
xoaytrai = nums[k:] + nums[:k]  # Lấy từ vị trí k đến cuối và nối với từ đầu đến vị trí k, cộng 2 dãy tuple lại
print(xoaytrai)
#Mở rộng ra xoay phải k vị trí
xoayphai= nums[-k:] + nums[:-k]  # Lấy từ vị trí -k đến cuối và nối với từ đầu đến vị trí -k, cộng 2 dãy tuple lại
print(xoayphai)
```

```python
#W6A4: Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp lại, thêm vào list của key đó.
dict_items = input().split()
result_dict = {} # Khởi tạo dictionary rỗng
for item in dict_items:
    key, value = item.split(':')
    if key not in result_dict:
        result_dict[key] = []  # Nếu key chưa có trong dictionary, khởi tạo với list rỗng
    result_dict[key].append(value)  # Thêm value vào list của key
print(result_dict)
```

```python
#W6A5: Nhập một list số nguyên. In ra dictionary {'positives': ..., 'negatives': ..., 'zeros': ...} (đếm số >0, <0, =0).
nums = list(map(int, input().split()))
result_dict = {'positives': 0, 'negatives': 0, 'zeros': 0}
for i in nums:
    if i > 0:
        result_dict['positives'] += 1 #Đếm số dương
    elif i < 0:
        result_dict['negatives'] += 1 #Đếm số âm
    else:
        result_dict['zeros'] += 1 #Đếm số =0
print(result_dict)
```

```python
#W6A7: Viết chương trình nhập vào một tuple, in ra phần tử đầu, phần tử cuối, và tuple đảo ngược.
tuple_input=tuple(map(int, input().split()))
print("Phần tử đầu:", tuple_input[0])
print("Phần tử cuối:", tuple_input[-1])
print("Tuple đảo ngược:", tuple_input[::-1])
#Cách 2 để in ra tuple đảo ngược
reversed_tuple = tuple(reversed(tuple_input))
print("Tuple đảo ngược (cách 2):", reversed_tuple)
```

```python
#W6A8: Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiện của từng chuỗi và lưu vào dictionary.
list_strings = input().split()
count_dict = {}
for s in list_strings:
    if s in count_dict: 
        count_dict[s] += 1 # Nếu chuỗi đã có trong dictionary, tăng số đếm lên 1
    else:
        count_dict[s] = 1 # Nếu chuỗi chưa có, khởi tạo số đếm là 1
print(count_dict)
```

```python
#W6A9: Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của key.
dict_1=eval(input())
dict_2=eval(input())
gop_dict = {}
for key in set(dict_1.keys()).union(set(dict_2.keys())):
    gop_dict[key] = dict_1.get(key, 0) + dict_2.get(key, 0) # Lấy giá trị từ dict_1 và dict_2, nếu key không tồn tại thì lấy giá trị mặc định là 0
# Sử dụng set union để lấy tất cả key từ cả hai dictionary, sau đó cộng giá trị nếu key tồn tại trong cả hai
for key in sorted(gop_dict.keys()):
    print(f"'{key}': {gop_dict[key]}") # In ra dictionary mới theo thứ tự từ điển (chữ cái) tăng dần của key
```

```python
#W6A10: Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.
nums=list(map(int, input().split()))
k=int(input())
pairs = set() # Sử dụng set để tránh trùng lặp
for i in range(len(nums)):
    for j in range(i + 1, len(nums)): # Đảm bảo thứ tự i < j
        if nums[i] + nums[j] == k: # Kiểm tra tổng bằng k
            pairs.add((nums[i], nums[j])) # Thêm cặp (nums[i], nums[j]) vào set
result = sorted(pairs) # Chuyển set thành list và sắp xếp
print(result)
```

```python
#W6A11: Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên thứ tự tương đối. In ra hai tuple này trên hai dòng riêng biệt.
tuple_input = tuple(map(int, input().split()))
even_numbers = tuple(x for x in tuple_input if x % 2 == 0)
odd_numbers = tuple(x for x in tuple_input if x % 2 != 0)
print(even_numbers)
print(odd_numbers)
```

```python
#W6A12: Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất trong số đó.
list_nums = list(map(int, input().split()))
count_dict = {}
for num in list_nums:
    if num in count_dict:
        count_dict[num] += 1
    else:
        count_dict[num] = 1 # Đếm số lần xuất hiện của từng số 
# Tìm số xuất hiện nhiều lần nhất
max_count = max(count_dict.values())
candidates = [num for num, count in count_dict.items() if count == max_count]
print(min(candidates))  # In ra số nhỏ nhất trong các số có số lần xuất hiện nhiều nhất
```

```python
#W6A13: Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo một dictionary mới bằng cách đảo ngược key và value của dictionary ban đầu. In ra dictionary mới.
dict_input = eval(input())
reversed_dict = {value: key for key, value in dict_input.items()} # Đảo ngược key và value, value ở mỗi key khác nhau
print(reversed_dict)  
```

```python
#W6A14: Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. Hãy tìm các phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các phần tử chung đó, không trùng lặp.
nums1 = list(map(int, input().split()))
nums2 = list(map(int, input().split()))
common_elements = list(set(nums1) & set(nums2)) # Tìm giao của hai tập hợp và chuyển thành list
print(common_elements)
```

```python
#W6A15: Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary mới.
dict_input = eval(input())
k = int(input())
filtered_dict = {key: value for key, value in dict_input.items() if value > k} # Lọc các cặp key-value có value > k
print(filtered_dict)
```
