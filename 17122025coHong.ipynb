```python
# 3.
def count_word_frequency(sentence):
    words = sentence.split() # Tách câu thành các từ dựa trên khoảng trắng
    frequency = {}
    
    for word in words:
            word = word.strip('.,!').lower() # Loại bỏ dấu câu và chuyển về chữ thường
        
        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1
    return frequency

text = "Học lập trình Python rất vui và Python rất mạnh mẽ"
result = count_word_frequency(text)
print(result)
```

```python
def count(n):
    count={}
    for i in n:
        if i in count:
            count[i]+=1
        else:
            count[i]=1
    return count
n= ["hello", "world", "hello", "world"]
print(count(n))
```

```python
students = {
    'S001': {'name': 'An', 'score': 27, 'major': 'CS'},
    'S002': {'name': 'Bình', 'score': 22, 'major': 'CT'},
}
students['S003'] = {'name': 'Cường', 'score': 21, 'major': 'IS'}
tong = 0
for student in students.values():
    tong += student['score']
trung_binh_cong = tong / len(students)
print(trung_binh_cong)
```

```python
students = {
    'S001': {'name': 'An', 'age': 27, 'major': 'CS'},
    'S002': {'name': 'Bình', 'age': 22, 'major': 'CT'},
}
students['S003'] = {'name': 'Cường', 'age': 21, 'major': 'IS'}
total_age = 0
for i in range(1, 4):  
    key = f'S00{i}'    
    total_age += students[key]['age']
    avg_age = total_age / len(students)
print(avg_age)
```
