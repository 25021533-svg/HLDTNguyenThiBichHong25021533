```python
#slide số 6/53: How to print a Countdown with Loop
def blast_off_loop(n):
    for n in range(n, 0, -1):
        print(n)
    print('BLAST OFF!')
n=int(input())
print(blast_off_loop(n))
```

```python
#slide số 6/53: How to print a Countdown with Loop
def blast_off_loop(n):
    if n==0:
        print('BLAST OFF!')
    else:
        print(n)
        blast_off_loop(n-1)
n=int(input())
print(blast_off_loop(n))
```

```python
#Slide số 12/53: The Factorial Function (Tính giai thừa)
def factorial(n):
    if n<0:
        return "Undefined for negative numbers"
    elif n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
n=int(input())
print(factorial(n))
```

```python
#Slide số 12/53: The Factorial Function (Tính giai thừa)
def factorial(n):
    result=1
    for i in range(1,n+1):
        result=result*i
    return result
```

```python
#Slide số 16/53: Tính tổng ngược từ n về 0 sử dụng đệ quy
def mystery(n):
    if n==0:
        return 0
    else:
        return n+mystery(n-1)
n=4
print(mystery(n))
```

```python
#Slide số 29/53: Dãy fibonacci (Số sau bằng tổng 2 số trước, ví dụ 5=3+2)
def fibo_rec(n):
    if n==0 or n==1:
        return n
    else:
        return fibo_rec(n-1)+fibo_rec(n-2)
n=int(input())
print(fibo_rec(n))
```
