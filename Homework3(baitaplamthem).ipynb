```python
# Bài 1: In ra các số dương lẻ bé hơn 100
for i in range(1,100,2):
    print(i)
# Sử dụng vòng lặp for với bước nhảy là 2 để in ra các số lẻ từ 1 đến 99.
```

```python
#Bài 2. In các số dương chẵn chia hết cho 3 bé hơn 100
i = 0
while i < 100:
    if i % 6 == 0:
        print(i, end=" ")
    i += 1

# Sử dụng vòng lặp while để duyệt các số từ 0 đến 99 và kiểm tra nếu số đó chia hết cho 6 (chẵn và chia hết cho 3) thì in ra.
```

```python
# Bài 3: Bảng cửu chương a
a = int(input("Nhập số a: "))
for i in range(1, 11):
    print(f"{a} x {i} = {a*i}")
# Sử dụng vòng lặp for để in bảng cửu chương của số a từ 1 đến 10.
```

```python
# Bài 4: IN tam giác cân rỗng cao h
h = int(input("Nhập h: "))
for i in range(1, h + 1):
    if i == 1:
        print(" " * (h - i) + "*")
    elif i < h:
        print(" " * (h - i) + "*" + " " * (2*i - 3) + "*")
    else:
        print("*" * (2*h - 1))

# Sử dụng vòng lặp for để in từng dòng của tam giác cân rỗng dựa trên chiều cao h nhập vào.
```

```python
# Bài 5: S= 1 + ... +n
n = int(input("Nhập n: "))
s = 0
i = 1
while i <= n:
    s += i
    i += 1
print("S =", s)
# Sử dụng vòng lặp while để tính tổng S từ 1 đến n.
```

```python
# Bài 6. In các ước dương của a
a = int(input("Nhập a: "))
i = 1
while i <= a:
    if a % i == 0:
        print(i, end=" ")
    i += 1
# Sử dụng vòng lặp while để tìm và in tất cả các ước dương của số a.
```

```python
#Bài 7. Đếm số ước dương của a
a = int(input("Nhập a: "))
count = 0
i = 1
while i <= a:
    if a % i == 0:
        count += 1
    i += 1
print("Số ước:", count)
```

```python
#Bài 8. In ước dương chung của a và b
a = int(input("a = "))
b = int(input("b = "))
i = 1
m = a if a < b else b
while i <= m:
    if a % i == 0 and b % i == 0:
        print(i, end=" ")
    i += 1
```

```python
#Bài 9. Kiểm tra số chính phương
a = int(input("Nhập a: "))
i = 1
cp = False
while i*i <= a:
    if i*i == a:
        cp = True
        break
    i += 1
print("Chính phương" if cp else "Không phải chính phương")
```

```python
#Bài 10. Kiểm tra số hoàn hảo
a = int(input("Nhập a: "))
s = 0
i = 1
while i < a:
    if a % i == 0:
        s += i
    i += 1
print("Hoàn hảo" if s == a else "Không hoàn hảo")
# Sử dụng vòng lặp while để tính tổng các ước dương của a (ngoại trừ chính nó) và kiểm tra nếu tổng này bằng a thì a là số hoàn hảo.
```

```python
#Bài 12. Tìm k lớn nhất sao cho S(k) < n
n = int(input("Nhập n: "))
s = 0
k = 0
while True:
    if s + (k + 1) < n:
        k += 1
        s += k
    else:
        break
print("k =", k)
# Sử dụng vòng lặp while để tìm k lớn nhất sao cho tổng S(k) nhỏ hơn n.
```

```python
#Bài 13. Tìm n nhỏ nhất sao cho 1 + 1/2 + ... + 1/n > A
A = float(input("Nhập A: "))
s = 0
n = 1
while True:
    s += 1/n
    if s > A:
        break
    n += 1
print("n =", n)
# Sử dụng vòng lặp while để tìm n nhỏ nhất sao cho tổng dãy 1 + 1/2 + ... + 1/n lớn hơn A.
```

```python
#Bài 14. Nhập dãy, dừng khi -1, tìm max và min
first = True
while True:
    x = int(input("Nhập số: "))
    if x == -1:
        break
    if first:
        mn = mx = x
        first = False
    else:
        if x > mx:
            mx = x
        if x < mn:
            mn = x
print("Max:", mx, "Min:", mn)
# Sử dụng vòng lặp while để nhập dãy số cho đến khi gặp -1, đồng thời cập nhật giá trị lớn nhất và nhỏ nhất trong dãy.
```

```python
#Bài 15. Đếm số chữ số của n Cách 1: chia 10
n = int(input("Nhập n: "))
count = 0
while n > 0:
    count += 1
    n //= 10
print(count)
# Sử dụng vòng lặp while để chia n cho 10 liên tục và đếm số lần chia cho đến khi n trở thành 0.
```

```python

#Bài 15. Cách 2: tăng mốc 10
n = int(input("Nhập n: "))
digits = 1
t = 10
while t <= n:
    digits += 1
    t *= 10
print(digits)
# Sử dụng vòng lặp while để nhân t với 10 liên tục và đếm số lần nhân cho đến khi t lớn hơn n.
```

```python
#Bài 16. Đếm chữ số chẵn và lẻ
n = int(input("Nhập n: "))
chan = le = 0
while n > 0:
    d = n % 10
    if d % 2 == 0:
        chan += 1
    else:
        le += 1
    n //= 10
print("Chẵn:", chan, "Lẻ:", le)
# Sử dụng vòng lặp while để lấy từng chữ số của n, kiểm tra tính chẵn lẻ và đếm số lượng chữ số chẵn và lẻ.
```

```python
# Bài 17. Tổng các chữ số của n
n = int(input("Nhập n: "))
s = 0
while n > 0:
    s += n % 10
    n //= 10
print("Tổng =", s)
# Sử dụng vòng lặp while để lấy từng chữ số của n, cộng dồn vào tổng s và in kết quả.
```

```python
# Bài 18. Kiểm tra số dạng 3^k
n = int(input("Nhập n: "))
if n < 1:
    print("Không")
else:
    ok = True
    while n > 1:
        if n % 3 != 0:
            ok = False
            break
        n //= 3
    print("Đúng dạng 3^k" if ok else "Không phải 3^k")
# Sử dụng vòng lặp while để chia n cho 3 liên tục và kiểm tra nếu n không chia hết cho 3 thì n không phải dạng 3^k.
```

```python
# Bài 19. Fibonacci lớn nhất không vượt quá A
A = int(input("Nhập A: "))
f1 = 1
f2 = 1
while True:
    f3 = f1 + f2
    if f3 > A:
        break
    f1 = f2
    f2 = f3
print(f2)
# Sử dụng vòng lặp while để tính dãy Fibonacci cho đến khi số Fibonacci tiếp theo vượt quá A, sau đó in số Fibonacci lớn nhất không vượt quá A.
```

```python
# Bài 20. Đếm chữ số nguyên tố trong n
n = int(input("Nhập n: "))
count = 0
while n > 0:
    d = n % 10
    if d in (2, 3, 5, 7):
        count += 1
    n //= 10
print("Số chữ số nguyên tố:", count)
# Sử dụng vòng lặp while để lấy từng chữ số của n, kiểm tra nếu chữ số đó là số nguyên tố (2, 3, 5, 7) và đếm số lượng chữ số nguyên tố.
```
