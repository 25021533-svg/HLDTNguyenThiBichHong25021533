```python
#Câu 1: Nhập vào một dãy số nguyên gồm n phần tử. Tìm và in ra giá trị lớn nhất trong dãy.
def max_value(n):
    max_val = n[0] #Giả sử phần tử đầu tiên là lớn nhất
    for i in n:
        if i > max_val:
            max_val = i
    return max_val
n=list(map(int, input().split()))
print(max_value(n))
```

```python
#Câu 2: Cho 2 danh sách số nguyên có vài phần tử trùng lặp. In ra các phần tử chung của hai danh sách đó không trùng lặp.
a=list(map(int, input().split()))
b=list(map(int, input().split()))
c=[]
for i in a:
    if i in b and i not in c: #kiểm tra phần tử i có trong b và chưa có trong c, nếu chưa có thì thêm vào c
        c.append(i)
print(list(c))
```

```python
#Câu 2: Cho 2 danh sách số nguyên có vài phần tử trùng lặp. In ra các phần tử chung của hai danh sách đó không trùng lặp.
a=list(map(int, input().split()))
b=list(map(int, input().split()))
c=[]
for i in a:
    if i in b: #kiểm tra phần tử i có trong b
        c.append(i)
c=set(c) #chuyển c thành tập hợp để loại bỏ phần tử trùng lặp
print(list(c))
```

```python
#Câu 3: Nhập vào hai số nguyên dương tạo thành một cặp số thân thiết khi chúng bằng tổng các ước số của nhau (không tính chính nó). Viết chương trình kiểm tra hai số có phải là cặp số thân thiết hay không.
def sum_of_divisors(num):
    total = 0
    for i in range(1, num):
        if num % i == 0:
            total += i
    return total
a, b = map(int, input().split())
if sum_of_divisors(a) == b and sum_of_divisors(b) == a:
    print(f"{a} và {b} là cặp số thân thiết")
```

```python
#Câu 4: Viết một số nguyên dương n. Xem nó có phải số hoàn hảo hay không.
def is_perfect_number(n):
    tong_uoc = 0
    for i in range(1, n):
        if n % i == 0:
            tong_uoc += i
    return tong_uoc == n #Biểu thức logic, nếu đúng thì true, sai thì false
n = int(input())
print(is_perfect_number(n))
```

```python
#Câu 6: Viết 1 số đổi sang giờ, phút, giây.
n=int(input("Nhập số giây: "))
gio=n//3600 #Số giờ là số giây chia cho 3600, lấy phần nguyên
phut=(n%3600)//60 #Số phút là phần dư của số giây chia cho 3600, chia tiếp cho 60, lấy phần nguyên
giay=n%60 #Số giây là phần dư của số giây chia cho 60
print(f"{gio} giờ {phut} phút {giay} giây")
```

```python
#Câu 6: Viết 1 số đổi sang giờ, phút, giây.
def convert_seconds(giay):
    gio = giay // 3600
    phut = (giay % 3600) // 60
    giay = giay % 60
    return gio, phut, giay
n = int(input("Nhập số giây: "))
print(convert_seconds(n))
```
