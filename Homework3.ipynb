```python
#W3A4 (Viết chương trình chia một số m cho n, kết quả thu được làm tròn xuống)
a=float(input("Nhập số m: "))
b=float(input("Nhập số n: "))
import math
if a>=0 and b >0:
    print("Kết quả làm tròn xuống:", math.floor(a/b))
elif a<0 and b>0:
    print("Kết quả làm tròn xuống:", math.ceil(a/b))
elif a>=0 and b<0:
    print("Kết quả làm tròn xuống:", math.ceil(a/b))
else:
    print("Kết quả làm tròn xuống:", math.floor(a/b))
# Chia 2 trường hợp: a, b âm và dương để làm tròn đúng hướng.
# Làm tròn xuống: Nếu cả 2 số cùng dấu thì dùng hàm floor, khác dấu dùng hàm ceil.
```

```python
#W3A6 (Nhập vào một số nguyên dương x, bạn hãy kiểm tra xem x là số chẵn hay lẻ.)
x=int(input("Nhập số nguyên x: "))
if x>0:
    if x%2==0:
        print(x,"là số chẵn")
    else:
        print(x,"là số lẻ")
else:
    print(x, "không phải số nguyên dương, vui lòng nhập lại")
# Kiểm tra số nguyên dương chẵn lẻ bằng phép chia lấy dư.
# Nếu số chia hết cho 2 thì là số chẵn, ngược lại là số lẻ.
```

```python
#W3A7 (Nhập vào hai số nguyên a, b. Nếu cả hai số a, b đều âm thì in ra "Yes", ngược lại in ra "No".)
a=int(input("Nhập số nguyên a: "))
b=int(input("Nhập số nguyên b: "))
if a<0 and b<0:
    print("Yes")
else:
    print("No")
# Sử dụng hàm if...else để kiểm tra điều kiện cả hai số âm.
```

```python
#W3A8 (Nhập vào hai chuỗi a, b gồm các chữ cái thường. In ra "True" nếu chuỗi a dài hơn chuỗi b. Ngược lại in ra "False".)
a=input("Nhập chuỗi a: ")
b=input("Nhập chuỗi b: ")
if len(a)>len(b):
    print("True")
else:
    print("False")
# Sử dụng hàm if...else (hàm điều kiện) + hàm len (hàm độ dài chuỗi kí tự) để so sánh độ dài của hai chuỗi.
```

```python
#W3A9 (Nhập vào 3 số nguyên dương a, b, c. Kiểm tra xem a, b, c có cấu thành độ dài cạnh của 1 tam giác được không. Nếu a, b, c cấu tạo thành được một tam giác, in ra "Yes". Ngược lại in ra "No".)
a,b,c=map(int,input("Nhập 3 cạnh của tam giác: ").split())
if a>0 and b>0 and c>0:
    if a+b<=c or a+c<=b or b+c<=a:
        print("Không tạo thành một tam giác")
    else:
        if a==c or a==b or b==c:
            print("Yes, tam giác cân")
        elif a==b and b==c:
            print("Yes, tam giác đều")  
        elif a**2+b**2==c**2 or b**2+c**2==a**2 or c**2+a**2==b**2:
            print("Yes, tam giác vuông")
        else:
            print("Yes, tam giác thường")
else: 
    print("Không thể tạo thành tam giác")
# Xét trường hợp một trong ba cạnh nhỏ hơn hoặc bằng 0 thì không thể tạo thành tam giác.
# Nếu ba cạnh đều lớn hơn 0 thì xét tiếp các trường hợp:
# + Nếu tổng hai cạnh bất kì nhỏ hơn hoặc bằng cạnh còn lại thì không thể tạo thành tam giác.
# + Nếu ba cạnh bằng nhau thì là tam giác đều.
# + Nếu hai cạnh bằng nhau thì là tam giác cân.
# + Nếu bình phương của một cạnh bằng tổng bình phương của hai cạnh còn lại thì là tam giác vuông.
# + Các trường hợp còn lại là tam giác thường.
```

```python
#W3A10 (Nhập vào 4 số nguyên. In ra màn hình số nguyên lớn nhất trong 4 số đó.)
a,b,c,d=map(int,input("Nhập 4 số nguyên: ").split())
max_num=a
if b>max_num:
    max_num=b
elif c>max_num:
    max_num=c
elif d>max_num:
    max_num=d
print("Số nguyên lớn nhất là:", max_num)
# Sử dụng cấu trúc if...elif để so sánh + hàm max_num (số lớn nhất) và tìm ra số lớn nhất trong 4 số nguyên nhập vào.
```
