```python
#Bài 2
a = float(input("Nhập chiều rộng a: "))
b = float(input("Nhập chiều dài b: "))
pi = 3.14
r = min(a, b) / 2
dien_tich_dat = a * b
dien_tich_hinh_tron = pi * (r ** 2)
dien_tich_trong_cay = dien_tich_dat - dien_tich_hinh_tron
print("Diện tích trồng cây là:", round(dien_tich_trong_cay, 2))
```

```python
c = input("Nhập một ký tự: ")
if c.isupper():
        print("Chữ thường tương ứng:", c.lower())
elif c.islower():
        print("Chữ hoa tương ứng:", c.upper())
else:
        print("Không phải chữ cái.")
```

```python
#Bài 5
c = input("Nhập chữ cái hoa (A-Z): ")
if c == 'A':
    print("Không có chữ cái thường liền trước a.")
elif 'A' < c <= 'Z':
    c_thuong = chr(ord(c.lower()) - 1)
    print(f"Chữ cái thường liền trước là: {c_thuong}")
else:
    print("Ký tự nhập vào không hợp lệ.")
```

```python
#Bài 6
import math
a, b, c = map(float, input("Nhap vao 3 so: ").split())
if a+b>c and b+c>a and a+c>b:
    p=(a+b+c)/2
    s=math.sqrt(p*(p-a)*(p-b)*(p-c))
    print(s)
else:
    print("Khong phai 3 canh cua tam giac")
```
