```python
# Khai báo biến trong Python - Nhập giá trị từ bàn phím và chuyển đổi kiểu dữ liệu
x = int(input("Nhập một số nguyên: "))  # Nhập giá trị từ bàn phím và chuyển đổi sang số nguyên
print("Giá trị của x là:", x)
y = float(input("Nhập một số thực: "))  # Nhập giá trị từ bàn phím và chuyển đổi sang số thực
print("Giá trị của y là:", y)
name = input("Nhập tên của bạn: ")  # Nhập giá trị từ bàn phím
print("Tên của bạn là:", name)
is_student = input("Bạn có phải là sinh viên không? (yes/no): ").lower() == 'yes'  # Nhập giá trị từ bàn phím và chuyển đổi sang boolean
print("Bạn là sinh viên:", is_student)
```

```python
#Chuyển đổi kiểu dữ liệu trong Python
a = "123"  # Biến chuỗi
print("Kiểu dữ liệu ban đầu của a:", type(a))  # In kiểu dữ liệu ban đầu
a_int = int(a)  # Chuyển đổi sang số nguyên
print("Giá trị của a sau khi chuyển đổi sang int:", a_int)
print("Kiểu dữ liệu của a sau khi chuyển đổi sang int:", type(a_int))  # In kiểu dữ liệu sau chuyển đổi
a_float = float(a)  # Chuyển đổi sang số thực
print("Giá trị của a sau khi chuyển đổi sang float:", a_float)
print("Kiểu dữ liệu của a sau khi chuyển đổi sang float:", type(a_float))  # In kiểu dữ liệu sau chuyển đổi
a_bool = bool(int(a))  # Chuyển đổi sang boolean
print("Giá trị của a sau khi chuyển đổi sang bool:", a_bool)
print("Kiểu dữ liệu của a sau khi chuyển đổi sang bool:", type(a_bool))  # In kiểu dữ liệu sau chuyển đổi
```

```python
#Chuỗi kiểu 02 trong Python - Truy xuất và cắt chuỗi
str="Tư duy tính toán"
print("Kí tự đầu tiên của str là: ", str[0]) #In kí tự đầu tiên
print("Kí tự cuối cùng của str là: ", str[-1]) #In kí tự cuối cùng
print("Kí tự đầu tiên đến kí tự thứ 4 là: ", str[0:4]) #In từ kí tự đầu đến kí tự thứ 4
print("Kí tự thứ 4 đến kí tự thứ 7 là: ", str[3:7]) #In từ kí tự thứ 4 đến kí tự thứ 7
print("Từ đầu đến kí tự thứ 5 là: ", str[:8]) #In từ đầu đến kí tự thứ 8
print("Từ kí tự 6 đến cuối là: ", str[7:]) #In từ kí tự thứ 8 đến cuối
print("Toàn bộ chuỗi là: ", str[:]) #In toàn bộ chuỗi
print("Toàn bộ chuỗi với bước nhảy là 2: ", str[::2]) #In toàn bộ chuỗi với bước nhảy là 2
print("Toàn bộ chuỗi in ngược lại là: ", str[::-1]) #In toàn bộ chuỗi ngược lại
```
